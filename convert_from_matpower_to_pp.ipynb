# Convert the 4-area network to pandapower 
This notebook reads in the 4 area network from a matpower file
![Four area test network](4-area/4_area.jpg)

## Load the network
Now we load pandapower and create the network

```python
import pandapower as pp
net=pp.converter.from_mpc("4-area/4_area.mat", f_hz=50, casename_mpc_file='mpc', validate_conversion=False)
```

Check if we can run a load flow

```python
pp.runpp(net, numba=True)
```

```python
net.res_line
```

Check if we can find the admittance matrix

```python
net["_ppc"]["internal"]["Ybus"]
```

## Save the network
Now we save the network in a pandapower compatible format.

```python
pp.to_json(net, "4-area/4_area.json")
```

## Load the network
Check if we can load the network

```python
net2 = pp.from_json("4-area/4_area.json")
pp.runpp(net2, numba=True)
```
